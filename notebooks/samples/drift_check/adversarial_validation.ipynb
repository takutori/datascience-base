```python
import numpy as  np
import sys
import os

# notebooks/ の1つ上の階層（=プロジェクトルート）をパスに追加
sys.path.append(os.path.abspath("../../../"))
```

```python
from dsbase.drift_check.adversarial_validation import AdversarialValidation, Vizalize, AdValLightGBMTuner
```

# 分布差がないケース

```python
train_X = np.random.randn(10000, 2)
test_X = np.random.randn(2000, 2)
```

```python
y_true, y_predict = AdversarialValidation(
    AdValLightGBMTuner(
        val_size=0.2,
        test_size=0.2
        )
).run(train_X, test_X)
```

```python
vizalize = Vizalize(
    y_true=y_true,
    y_predict=y_predict
)
```

```python
fig, roc_auc = vizalize.roc_curve()
```

## 分布差があるケース

```python
train_X = np.random.randn(10000, 2)*5
test_X = np.random.randn(2000, 2)*5 + 4
```

```python
y_true, y_predict = AdversarialValidation(
    AdValLightGBMTuner(
        val_size=0.2,
        test_size=0.2
        )
).run(train_X, test_X)
```

```python
vizalize = Vizalize(
    y_true=y_true,
    y_predict=y_predict
)
```

```python
fig, roc_auc = vizalize.roc_curve()
```
